**Parts of BDA**
1. Univariate Analysis=>>>Analysis on single Independent column
2. Bivariate Analysis=>>> Analysis on two columns
3. Multivariate Analysis=>>> Analysis on more than 2 columns

```python
# Data Types
# 1. Numerical Data =>> Continuous Data =>> Age(year,month,days), Height , weight etc..
# 2. Categorical Data=>>> Decribe Data ==>>> Total Number of Employees
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Univariate Analysis**

```python
df.columns
```

```python
sns.countplot(x='Survived',data=df)
```

```python
# If we want to see actual values then we use value_counts().

df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
plt.hist(x=df['Age'])
```

```python
# 2.Distplot
# 3.curve==>>> KDE(Kernal Density Exactraction)
# use for find curve

sns.distplot(df['Age'])
```

```python
# 3. Boxplot
# for find our outliers

# 1. lower fence
# 2. 25% data
# 3. IQR(Inter Quartile Range)(75%-25%)
# 4. 75% data
# 5. upper fence

sns.boxplot(x='Age',data=df)
```
